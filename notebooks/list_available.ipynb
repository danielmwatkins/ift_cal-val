The IFT pipeline only is able to run a subset of the images. Here, we find the overlap between the images where floes are visible and where IFT successfully segmented the image.

```python
# get list of evaluation tables
import pandas as pd
import os
eval_table_loc = "../data/ift_data/ift_pipeline_minarea_100px/eval_tables/"
files = os.listdir(eval_table_loc)

assign_table_loc = "../data/validation_tables/floe_labeling_tables/"
afiles = os.listdir(assign_table_loc)

df_eval = pd.concat([pd.read_csv(eval_table_loc + file) for file in files])
df_assign = pd.concat([pd.read_csv(assign_table_loc + file, index_col=0) for file in afiles if '.csv' in file]) 
```

```python
df_assign['location'] = [x.replace(x.split('-')[0] + '-', '') for x in df_assign['long_name']]
```

```python
df = df_assign.merge(df_eval[['location', 'preprocess', 'tracker']], left_on='location', right_on='location', how='inner')
```
